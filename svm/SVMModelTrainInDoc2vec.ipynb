```python
##读取向量
def getVecs(model, corpus):
    #vecs = [np.array(model.wv[z.strip()] if z.strip() in model.wv else np.zeros(100) ) for z in corpus]
    vecs = [np.array(model.infer_vector([z.strip()] , alpha=0.05, steps=iterNum)) for z in corpus ]
    return vecs
```
